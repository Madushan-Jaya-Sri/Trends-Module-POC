### Related topics

```python
from serpapi import GoogleSearch
api_key = os.getenv("SERPAPI_API_KEY")

params = {
  "engine": "google_trends",
  "q": "coffee",
  "geo": "LK",
  "data_type": "RELATED_TOPICS",
  "api_key": api_key
}

search = GoogleSearch(params)
results = search.get_dict()
related_topics = results["related_topics"]
```

```python
related_topics
```

### Related queries
